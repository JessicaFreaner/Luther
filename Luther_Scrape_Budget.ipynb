```python
import pprint as pp
import urllib2
from bs4 import BeautifulSoup
import dateutil.parser
import re
import pandas as pd
import numpy as np

import pickle
```

```python
# extract soup for urlname
def url_to_soup(urlname):
    try:
        page = urllib2.urlopen(urlname)
        soup = BeautifulSoup(page)
    except:
        print urlname
        soup = 'N/A'
        print 'soup', soup
    return soup
```

```python
def convert_raw_row_to_clean_data(raw_data, skip_row=[], cut_off_rows=0):
    clean_row_data = []
    headers = []
    no_of_rows = len(raw_data)
    for count, row in enumerate(raw_data):
        cols = row.find_all('td')
        cols = [ele.text.strip() for ele in cols]

        if count in skip_row or count >= (no_of_rows - cut_off_rows):
            continue
        elif count == 0: # can be header row or skipped by user
            headers = [ele.replace(' ', '_') for ele in cols if ele] # get rid of empty values
        else:
            clean_row_data.append([ele for ele in cols if ele]) # get rid of empty values
    if len(set(headers)) != len(headers):
        headers = []
            
    return headers, clean_row_data
```

```python
def make_df(row_data, headers):
    try:
        df = pd.DataFrame(row_data, columns=headers)
    except:
        df = pd.DataFrame(row_data)
    
    return df
```

```python
# extract and clean data from box office mojo seasonal URL
def extract_table_data(urlname, table_number, skip_row_lst=[], cut_off_rows=0):
    soup = url_to_soup(urlname)
    table = soup.find('table', attrs={'class':'lineItemsTable'})     # finds tables in html

    # extracts raw data from url page
    raw_row_data = soup.find_all('table')[table_number].find_all('tr')
    headers, clean_row_data = convert_raw_row_to_clean_data(raw_row_data, skip_row_lst, cut_off_rows)

    df = make_df(clean_row_data, headers)

    return df
```

```python
def drop_excess_data(df, page='Season'):
    if page == 'Season':
        df.drop(df.columns[[1,2,3,4,5,6,7,9,10]], axis=1, inplace=True)
    else:
        try:
            df.drop(df.columns[[0,2,3,4,5,6,7,8]], axis=1, inplace=True)
        except:
            df.drop(df.columns[[0,2,3,4,5,6,7]], axis=1, inplace=True)
    return df
```

```python
def add_seasonal_headers(df):
    df.columns = ['Year', 'Movie']
#     df.columns = ['Movie']
```

```python
def find_urls(soup, identifier):
    try:
        return soup.find_all('a', href=re.compile(identifier))
    except:
        print 'failed soup', soup
```

```python
def extract_season_urls(urlname):
    soup = url_to_soup(urlname)
    season_urls = []
    
    # identify url to other seasons
    link_list = find_urls(soup, '\?chart=byseason&season')
    for link in link_list[:4]:
        one_link = "http://boxofficemojo.com/seasonal/" + link.get('href')
        season_urls.append(str(one_link))
#     pp.pprint(season_urls)
    return season_urls
```

```python
def extract_movie_urls(urlname):
    
    soup = url_to_soup(urlname)
    movie_urls = []
    
    # identify movie urls
    link_list = find_urls(soup, '/movies/')
    last_pos = len(link_list) - 1
    excluded_link_pos = [0, 1, last_pos]
    
    for count, link in enumerate(link_list):
        if count in excluded_link_pos:
            continue
            
        # adjust values to 2015 dollars
        one_link = "http://boxofficemojo.com" + link.get('href') + "&adjust_yr=2015&p=.htm"
        try:
            movie_urls.append(str(one_link))
        except:
#             print 'failed on link', one_link
            one_link = u'%A0'.join(one_link.split())
#             print 'fixed link', one_link
            movie_urls.append(str(one_link))

    
    return movie_urls
```

```python
def extract_urls(urlname, skip_rows=0):
    
    movie_urls, season_urls, year_urls = [], [], []
    
    movie_urls = extract_movie_urls(urlname)
    season_urls = extract_season_urls(urlname)
#     year_urls = extract_year_urls(urlname)

#     return movie_urls[skip_rows:], season_urls, year_urls[skip_rows:]
    return movie_urls[skip_rows:], season_urls
```

```python
def get_movie_value(soup, field_name):
    obj = soup.find(text=re.compile(field_name))
    if not obj:
        return None
    
    next_sibling = obj.findNextSibling() # find sibling after search for string
    if next_sibling:
        return next_sibling.text
    else:
        return None
```

```python
def extract_movie_details(urlname):
    soup = url_to_soup(urlname)

    try:
        # budget
        budget = int(get_movie_value(soup, 'Production Budget').split(' ')[0].replace('$', '').replace(',',''))
    except:
        budget = None

    return budget
```

```python
def extract_movie_details_for_list(url_list):
    movie_specs_list = []
    for url in url_list:
        movie_specs_list.append(extract_movie_details(url))
    return movie_specs_list
```

```python
def concat_details_to_movie(df, movie_specs, season='Winter'):
    movie_specs_df = make_df(movie_specs,['Budget'])
    df = pd.concat([df, movie_specs_df], axis=1)
    df['Season'] = season
    return df
```

```python
def extract_data_for_season(url, season):
    df = extract_table_data(url, table_number=3)
    drop_excess_data(df)
    add_seasonal_headers(df)
    movie_urls, __ = extract_urls(url)
    movie_details = extract_movie_details_for_list(movie_urls)
    df = concat_details_to_movie(df, movie_details, season)
    
    return df
```

```python
def seasonal_top_movie_data_extraction():
    start_url = "http://boxofficemojo.com/seasonal/?view=widedate&adjust_yr=2015&p=.htm"

    # extract & clean main page date
    top_movies_df = extract_table_data(start_url, table_number=3, skip_row_lst=[1])
    drop_excess_data(top_movies_df)
    add_seasonal_headers(top_movies_df)

    # main page urls
    movie_urls, season_urls = extract_urls(start_url, skip_rows=1)

    # get dtg, genre, rating for main page data & join to df
    movie_details = extract_movie_details_for_list(movie_urls)
    top_movies_df = concat_details_to_movie(top_movies_df, movie_details)

    # get dtg, genre, rating for other seasons top movies & join to df
    seasons = ['Spring', 'Summer', 'Fall', 'Holiday']
    for count, url in enumerate(season_urls):
        other_top_movies_df = extract_data_for_season(url, seasons[count])
        top_movies_df = pd.concat([top_movies_df, other_top_movies_df])
#         year_urls += more_year_urls

    with open('top_movies_budget.pkl', 'w') as picklefile:
        pickle.dump(top_movies_df, picklefile)
#     # test for few years first
#     for url in year_urls:
#         other_movies_by_year_df = extract_other_movies_for_year(url)
#         all_movies_df = pd.concat([top_movies_df, other_movies_by_year_df])
    
    
#     # add dummy values for rating & genre
#     seasonal_movie_df = seasonal_movie_df.reset_index(drop=True) # reset indices
#     seasonal_movie_df = add_dummies_for_split_col(seasonal_movie_df, 'Genre')
# #     seasonal_movie_df = add_dummy_ratings(seasonal_movie_df, 'Rating')
    
    return top_movies_df
```

```python
def extract_other_movies_for_year(url, season='Winter'):
    # get year from url
    year = url.split('yr=')[1].split('&')[0]
    
    # skip headers, top movie, & summary
    other_movies_df = extract_table_data(url, table_number=3, skip_row_lst=[0, 1, 2], cut_off_rows=4) 
    drop_excess_data(other_movies_df, page='Year')
    # add year to dataframe
    other_movies_df.insert(0, 'Year', year)
    add_seasonal_headers(other_movies_df)
    movie_urls, __= extract_urls(url, skip_rows=1)
#     pp.pprint(movie_urls)
        
    # get budget & join to df
    movie_details = extract_movie_details_for_list(movie_urls)
    other_movies_df = concat_details_to_movie(other_movies_df, movie_details, season=season)

    return other_movies_df
```

```python
def other_movies_by_season(url_lst, years_in_season=33):

    # make df with first url
    season = url_lst[0].split('season=')[1] # get season from url first url
    print 'season', season
    other_movies_by_season_df = extract_other_movies_for_year(url_lst[0], 
                                                              season=season)

    # concat other year data to first url
    for url in url_lst[1:years_in_season]:
        other_movies_by_season_df = pd.concat([other_movies_by_season_df, 
                                               extract_other_movies_for_year(url, season=season)])
    with open('other_movies_budget_%s.pkl' %season, 'w') as picklefile:
        pickle.dump(other_movies_by_season_df, picklefile)
    
    return url_lst[years_in_season:]
```
