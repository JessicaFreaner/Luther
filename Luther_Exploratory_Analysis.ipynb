```python
def merge_dfs(left_df, right_df, on=None):
    return pd.merge(left_df, right_df, on=on)
    
```

```python
def clean_df(df, condense=False):
    clean_movies = df.dropna()  # removes 4 films
    
    clean_movies = clean_movies.reset_index(drop=True) # reset indices
    
    if condense:
        clean_movies = condense_rating(clean_movies)
        clean_movies = condense_genres(clean_movies)
        clean_movies = condense_decades(clean_movies)
        clean_movies = add_dummies(clean_movies, 'Season')
    
    return clean_movies
```

```python
def fix_budgets(x):
   if x > 1000:
       return x/1000000
   else:
       return x
```

```python
def adj_budget_for_inflation(df, dict=inflation_dict):
    df['Adj_Budget'] = 0
    for count, budget in enumerate(df['Clean_Budget']):
        if int(df.loc[count, 'Year']) in inflation_dict.keys():
            df.loc[count, 'Adj_Budget'] = df.loc[count,'Clean_Budget'] * inflation_dict[int(df.loc[count, 'Year'])]
    return df
```

```python
merged = merge_dfs(top_movies_df, top_movie_budgets_df, on=['Year', 'Movie', 'Season'])
clean_top_movies_df_w_budget = clean_df(merged)
clean_top_movies_df_w_budget['Clean_Budget'] = clean_top_movies_df_w_budget['Budget'].apply(fix_budgets)
adj_clean_top_movies_df_w_budget = adj_budget_for_inflation(clean_top_movies_df_w_budget)
```

```python
# apply() method lets you apply funciotn to DF/ series or portion of DF
def convert_year_to_decade(year):
    # test_year = 1982
    return str(year)[:3] + '0s'
```

```python
def add_decade_dummies(df, colname, name='Decade'):
    return pd.concat([df, pd.get_dummies(df[colname])], axis=1) 
```

```python
top_movies_df['Decade'] = top_movies_df['Year'].apply(lambda x: convert_year_to_decade(x))
top_movies_df = add_decade_dummies(top_movies_df, 'Decade')
# top_movies_df

all_wide_release_movies_df['Decade'] = all_wide_release_movies_df['Year'].apply(lambda x: convert_year_to_decade(x))
all_wide_release_movies_df = add_decade_dummies(all_wide_release_movies_df, 'Decade')
```
