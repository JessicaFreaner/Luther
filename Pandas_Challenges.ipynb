# Challenge 4: make one figure with (N=the number of MPAA ratings there are) subplots, and in each plot the release date vs the domestic total gross

```python
# matplotlib method to plot

plt.figure(figsize=(15,15))
sub = [4,3,2,1] # to identify subplot
color_dot = ['ro', 'go', 'bo', 'ko']
for rating, group in rating_groups:
    a = 220+sub.pop()
    plt.subplot(a)
    plt.plot(group.ReleaseDate, group.DomesticTotalGross, color_dot.pop())    
    plt.title(rating)
```

```python
# pandas method to plot

fig, axis_list = plt.subplots(nrows=2, ncols=2)
fig.set_size_inches(15,15)

for i, (rating, group) in enumerate(rating_groups):
    if i >1:
        row = 1
    else:
        row = 0 
    column = i % 2
    group.plot(x = 'ReleaseDate', y='DomesticTotalGross', ax=axis_list[row,column], title=rating, legend=False)
    
    
    
# rating_groups.get_group('PG').plot(x = 'ReleaseDate', y='DomesticTotalGross', ax=axis_list[0,0]), plt.legend(loc='best')
```

# Challenge 5: what director in your dataset has the highest gross per movie?

```python
directors = df.groupby('Director')
director_avg_gross = directors['Director', 'DomesticTotalGross'].mean()

director_avg_gross.sort('DomesticTotalGross', inplace=True, ascending=False)
director_avg_gross.head()
```

# Challenge 6: bin your dataset into years (if applicable) and make a bar graph with error bars of gross each year

```python
# apply() method lets you apply funciotn to DF/ series or portion of DF
df['ReleaseMonth'] = df['ReleaseDate'].apply(lambda x: x.month)
# df.head()
```

```python
release_month = df.groupby('ReleaseMonth')
month_avg_gross = release_month['DomesticTotalGross'].mean()
month_avg_gross_error = release_month['ReleaseMonth', 'DomesticTotalGross'].std()
```

```python
month_avg_gross
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(15,15)

# months = ['January', 'Feburary', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October',
#           'November', 'December']
month_avg_gross.plot(yerr=month_avg_gross_error, ax=ax, kind='bar', title='Domestic Total Gross by Month')

ax.set_ylabel('Total Gross (Millions of Dollars)')
```
